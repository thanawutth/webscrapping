```python
import bs4
import requests
```

```python
r = requests.get('https://www.tmd.go.th/province.php?StationNumber=48303')
```

```python
r.text
```

```python
html_page = bs4.BeautifulSoup(r.text, 'html.parser')
print(html_page)
```

```python
print(html_page.prettify())
```

```python
# titles = html_page.find_all('a')
titles = html_page.findAll('td', {'class':'MaxTemp'} )
print(titles)
```

```python
title_list = []
for t in titles:
    
    # Create a new variable --> obj to store 
    # only course name getting rid of unwanted tags
    obj = t.string
    
    # Append each course into a course_list variable
    title_list.append(obj)
```

```python
print(title_list)
```
